#  Line Graphs

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

 ### Straight Line

```python
plt.figure(figsize=(5, 4), dpi=150)
plt.title('Straight Line', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.style.use('ggplot')

plt.xlabel('Time')
plt.ylabel('Velocity')

plt.xticks([1, 2, 3, 4, 5])
plt.yticks([1, 2, 3, 4, 5])

x = [1, 2, 3, 4, 5]
y = [1, 2, 3, 4, 5]

# Shorthand notation
# fmt = '[color][marker][linestyle]'
# plt.plot(x, y, label='x = y', color='red')
plt.plot(x, y, color='purple', label='x = y')

plt.legend()

# save the graph
plt.savefig('StraightLine.jpg', dpi=(300))

plt.show()
```

### Pair of lines

```python
plt.figure(figsize=(5, 4), dpi=150)
plt.title('Compairing Two Lines', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.style.use('ggplot')

plt.xlabel('Time')
plt.ylabel('Velocity')

plt.xticks([1, 2, 3, 4, 5, 6])
plt.yticks([3, 6, 9, 12, 15, 18])

x1 = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
plt.plot(x1, y1, label='y = x²', color='red', linewidth=2, marker='.', markersize=10, linestyle='--', markeredgecolor='green')

x2 = [1, 2, 3, 4]
y2 = [1, 3, 5, 7]
plt.plot(x2, y2, label='y = 2x-1', color='blue', linewidth=2, marker='x', markersize=10, markeredgecolor='green')

plt.legend()

# save the graph
plt.savefig('PairOfLines.jpg', dpi=(300))

plt.show()
```

### Graph Slicing

```python
# Resize the graph
plt.figure(figsize=(5, 3), dpi=150)
plt.title('Two Different Line Graphs', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.style.use('ggplot')

x = np.arange(0, 4.5, 0.5)

plt.plot(x[:5], x[:5]**3, label='y=x³', color='green', marker='p')
plt.plot(x[4:], x[4:]**3, label='y=x³', color='purple', marker='o', linestyle='--')

plt.legend()

# save the graph
plt.savefig('GraphSlicing.jpg', dpi=(300))

plt.show()
```
