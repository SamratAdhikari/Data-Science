# Bar Graphs


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Bar Graph

```python
plt.figure(figsize=(5, 4), dpi=150)
plt.title('Bar Graph')
plt.style.use('ggplot')

labels = ['A', 'B', 'C']
values = [1, 4, 2]

bars = plt.bar(labels, values, color='#9136cf')

pattern = ['/', 'O', '*']

for bar in bars:
    bar.set_hatch(pattern.pop(0))


plt.ylabel('Census')
plt.savefig('BarGraph.jpg', dpi=300)

plt.show()

```
