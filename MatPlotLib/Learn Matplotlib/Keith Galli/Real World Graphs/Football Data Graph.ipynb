# Football Data Graph

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
fifa = pd.read_csv('football_data.csv')
# fifa.head(5)
```

### Histogram

```python
myBins = [40, 50, 60, 70, 80, 90, 100]

plt.style.use('seaborn')
plt.figure(dpi=150)
plt.hist(fifa.Overall, bins=myBins, color='purple')

plt.xticks(myBins)
plt.xlabel('Skill Level')
plt.ylabel('Number of Players')

plt.title('Distribution of Player Skills', fontdict={'fontname':'Comic Sans MS', 'fontweight':'bold'})
plt.savefig('Distribution of Player Skills.jpg', dpi=300)

plt.show()
```

### Pie Chart

```python
plt.figure(figsize=(7, 5), dpi=150)
plt.title('Foot Preference', fontdict={'fontname':'Comic Sans MS', 'fontweight':'bold', 'fontsize':20})
plt.style.use('ggplot')

left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

myLabels = ['Left', 'Right']

plt.pie([left, right], labels=myLabels, autopct='%.2f %%')

plt.savefig('Foot Preference of Players.jpg', dpi=300)
plt.show()
```
