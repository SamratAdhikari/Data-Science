# Gas Price

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Line Graph

```python
plt.figure(figsize=(7, 5), dpi=150)
plt.title('Gas Price over Time (in USD)', fontdict={'fontweight': 'bold', 'fontname':'Comic Sans MS'})
plt.style.use('seaborn-dark')

gas = pd.read_csv('gas_data.csv')
# gas

# plt.plot(gas.Year, gas.USA, label='United States', color='blue', marker='.')
# plt.plot(gas.Year, gas.Canada, label='Canada', color='red', marker='.')
# plt.plot(gas.Year, gas['South Korea'], label='South Korea', color='purple', marker='.')
# plt.plot(gas.Year, gas.Japan, label='Japan', color='orange', marker='.')

# shortcut for all countries
for country in gas:
    if country != 'Year':
        plt.plot(gas.Year, gas[country], label=str(country))

plt.xticks(gas.Year[::3].tolist()+[2011])
plt.xlabel('Year')
plt.ylabel('Price (in USD)')

plt.legend()
plt.savefig('Gas Price.jpg', dpi=300)

plt.show()
```
