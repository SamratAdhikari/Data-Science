### Stacking & Splitting

### Statistics

```python
stats = np.arange(1, 16, 2)
print(stats)

# mean
np.mean(stats)

# median
np.median(stats)

# average
np.average(stats)

# standard deviation
np.std(stats)

# variance
np.var(stats)

# if there is nan values in the array (non integer values), use these to ignore them
# nanmedian, nanmean, nanstd, nanvar

# using the csv file for statistics
np.percentile(iceSales, 50, axis=0) # 50% percentile of each column

# correlation coefficient
np.corrcoef(iceSales[:,0], iceSales[:,1]) # first column(temp), second column(sales)
# since the data are 1 or close to 1, temp and sales are highly co-related (directly proportional)
# this means when temp is high(hot), more icecreams are sold

# regression line
# Σ(x-x̅)*(y-ȳ) / Σ(x-x̅)^2
temp_mean = np.mean(iceSales[:,0])
sales_mean = np.mean(iceSales[:,1])

numerator = np.sum( (iceSales[:,0] - temp_mean) * (iceSales[:,1] - sales_mean) )
denominator = np.sum( np.square( iceSales[:,0] - temp_mean ) )

result = numerator / denominator

y_intercept = sales_mean - result * temp_mean
y_intercept

regArray = iceSales[:,0] * result + y_intercept
regArray
```

### Matrices & Linear Algebra

```python
Array1 = np.array( [[1, 2], [3, 4]] )
Array2 = np.array( [[2, 4], [6, 9]] )
Array3 = np.array( [[5, 6], [7, 8]] )

# print("Array 1:\n", Array1)
# print("Array 2:\n", Array2, '\n')


# -----------------------Matrix multiplication with Dot Product------------------------------
# (1 * 2) + (2 * 6) = 14 [0,0]
# (1 * 4) + (2 * 9) = 22 [0,1]
# (2 * 2) + (4 * 6) = 28 [1,0]
# (2 * 4) + (4 * 9) = 44 [1,1]
np.dot(Array1, Array2)                     # dot product between two matrices
la.multi_dot( (Array1, Array2, Array3) )   # dot product between more than two matrices


# ---------------------------------Inner product------------------------------------------------ 
# (similar to dot product but we multiply a row of an array to the row of another array)
# (1 * 2) + (2 * 4) = 10 [0,0]
# (1 * 6) + (2 * 9) = 24 [0,1]
# (2 * 2) + (4 * 4) = 22 [1,0]
# (2 * 6) + (4 * 9) = 54 [1,1]
np.inner(Array1, Array2)


# -------------------------------Tensor Dot Product-----------------------------------------------
# (1 * 1) + (2 * 2) + (3 * 3) + (4 * 4) = 30
# (5 * 1) + (6 * 2) + (7 * 3) + (8 * 4) = 70
Array4 = np.array( [ [[1, 2],
                      [3, 4]],
  
                    [[5, 6],
                      [7, 8]]
                   ] )
Array5 = np.array( [ [1, 2], [3, 4] ], dtype=object )

np.tensordot(Array4, Array5)


# ----------------------------------Einstein Summation----------------------------------------------- 
# Provides many ways to perform operations on multiple arrays
Array6 = np.array([0, 1])
Array7 = np.array( [ [0, 1, 2, 3], [4, 5, 6, 7] ])

# 'i,ij -> i' 
# from the LHS components from '->', i means we will be using one axis for array6 and ij means two axes for array7
# from RHS, i means we want one dimensional array
# ij : Means multiply array6 single item by each column of array7 and sum
# [0*4 + 0*5 + 0*6 + 0*7, 1*4 + 1*5 + 1*6 + 1*7] = [0, 4+5+6+7] = [0, 22]
np.einsum( 'i,ij -> i' , Array6, Array7)

np.einsum('i->' , Array6)  # add all the elements in Array6 (we used i only as array6 has only one dimension)
np.einsum('ij->', Array7)  # add all the elements from each row and add the results 
# [0+1+2+3] + [4+5+6+7] == 28 (we used ij because there array7 is 2 dimensional)

# dot product
Array8 = np.array([1, 2, 3, 4])
Array9 = np.array([2, 4, 6, 8])
np.einsum('i,i->', Array8, Array9) # 'i,i' because these two arrays are one dimensional

# matrix multiplication
# print("Array 1:\n", Array1)
# print("Array 2:\n", Array2, '\n')
np.einsum('ij,jk->', Array1, Array2)

# diagonal sum
np.einsum('ii', Array1) # gives sum of diagonal elements of array1

# transpose
np.einsum('ji', Array1)


# --------------------------Raise matrix to the power of n--------------------------------
# Given [[a, b], [c, d]]
# [[a² + bc, ab +db], [ac + dc, d² + bc]]
# must be multi dimensional array
la.matrix_power(Array1, 2) # array1 ^ 2


# --------------------------Kronecker product of 2 arrays----------------------------------
# Given [[a, b], [c, d]], [[e, f], [g, h]]
# [[a*e, a*f, b*e, b*f], [a*g, a*h, b*g, b*h], ...]
np.kron(Array1, Array2)


# ---------------------------Computer eigenvalues--------------------------------------
la.eig(Array1)     # returns eignen vectors
la.eigvals(Array1) # doesnt return eigen vectors


# ------------------------------Get Vector Norm-----------------------------------------
#  sqrt(sum(x**2)) # sum of square of each elements and then square root at the end
la.norm([3, 4])    # basically getting the value of hypotenuse
la.norm(Array1)


# -----------------------Multiplicative inverse of a matrix-------------------------------
la.inv(Array1)


# ---------------------------Condition number of matrix----------------------------------
la.cond(Array1)


# -----------------------------------Determinants---------------------------------------
# Determinates are used to compute volume, area, to solve systems of equations and more. 
# It is a way you can multiply values in a matrix to get 1 number.
# For a matrix to have an inverse its determinate must not equal 0
# det([[a, b], [c, d]]) = a*d - b*c
Array10 = np.array( [ [1, 2], [3, 4] ])
la.det(Array10) # 1*4 - 2*3 = -2

# determinant of 3x3 matrix
# det([[a, b, c], [d, e, f], [g, h, i]]) = a*e*i - b*d*i + c*d*h - a*f*h + b*f*g - c*e*g
# When we multiply a matrix times its inverse we get the identity matrix [[1,0],[0,1]] for a 2x2 matrix
# Calculate the inverse 1/(a*d - b*c) * [[d, -b], [-c, a]]
# 1/(4 - 6) = -.5 -> [[-.5*4, -.5*-2], [-.5*-3, -.5*a]]
Array10_inverse = la.inv(Array10)
np.dot( Array10, Array10_inverse )


# --------------------------------Solving Systems of Linear Equations------------------------------
# If you have 3x + 5 = 9x -> 5 = 6x -> x = 5/6
# Example:
# If you have x + 4y = 10 & 6x + 18y = 42
# Isolate x -> x = 10 - 4y
# 6(10 - 4y) + 18y = 42 -> 60 - 24y + 18y = 42 - > -6y = -18 -> y = 3
# x + 4*3 = 10 -> x = -2
Array11 = np.array( [ [1, 4], [6, 18] ] )
# from above example, [1, 4] are from LHS of 1st eqn and [6, 18] are from LHS of 2nd eqn
Array12 = np.array( [10, 42] )
# [10, 42] are from RHS of 1st and 2nd eqns respectively
la.solve(Array11, Array12)


# ---------------Return a identity matrix with defined number of rows and columns------------------------
np.eye(2, 1, dtype=int)
np.eye(3, 3, dtype=int)
```

### Financial Functions

```python
# Compute future value of 400 dollars investment every single month with an annual rate of 8%
# and see how much money i would have after 10 years

npf.fv(0.08/12, 10*12, -400, -400)
# 0.08/12 -> 8% per year --> (8/100) / 12 months ---> 0.08/12
# 10*12   -> result after 10 years --> 10*12 months
# -400    -> initial investment
# -400    -> 400 dollars montly investment after initial investment (installment)


# Calculate Compound Interest
# I wanna get a loan of $ 3000 from a bank, and they are loaning the money at an interest of 9.25% per year
# but thats gonna be compounded monthly and I wanna pay the loan with interest in 1 year period

period = np.arange(1*12) + 1 # create an array containing item from 1 to 12
                             # we added 1 because arrays start from 0 index and this array contains elements from 0 to 11
principle = 3000.00
interest_payment = npf.ipmt(0.0925/12, period, 1*12, principle)
principle_payment = npf.ppmt(0.0925/12, period, 1*12, principle)
# ipmt = interest payment
# ppmt = principle payment
# 0.0925/12 = 9.25% per year = (9.25/100) / 12 months
# period = i have to pay the money each month for a year (from 1st month to 12th month)
# 1*12 = i wanna pay the total sum in 1 year, 1 year = 12 month 
# principle = amount taken from the bank

# generate table
for index, payment in enumerate(period):
    principle = principle + principle_payment[index]
    
    print(f"{payment} : {np.round(principle_payment[index], 2)} : {np.round(interest_payment[index], 2)} : {np.round(principle, 2)}")
    # in round method, 2 denotes the 2 decimal places after the decimal (.)
    
    
# Calculate no. of payments to pay off $ 3000 if I decided  that i can pay a $ 150 per month 
# with an interest rate of 9.25% per month for a year
np.round(npf.nper(0.0925/12, -150, 3000.00), 2)


# Calculate net present value of cash flows. In the future, i expect to get 4000, 5000, 6000 and 7000 dollars 
# after initial 15000 dollar investment with added 8% interest rate

npf.npv(0.08, [-15000, 4000, 5000, 6000, 7000])
# 0.08 = interest rate
# -15000 = we took a loan of 15000
# we expect to earn 4000, 5000, 6000, 7000 dollars from that loan
# now we calculate net present value
```
