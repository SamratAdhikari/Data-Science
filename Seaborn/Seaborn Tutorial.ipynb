## Types of Distribution Plots

Distribution plots visually assess the distribution of sample data by comparing the empirical 
(experimental) distribution of the data with the theoretical values expected from a specified distribution.

#### Distribution (Dist) Plot

- It is used basically for univariant set of observations and visualizes it through a histogram   
    i.e. only one observation and hence we choose one particular column of the dataset.

```python
sns.distplot(crash['not_distracted'], bins=25) # kde=False to remove the normal dist line
plt.show()
```

#### Joint Plot

- Joint Plot is used to quickly visualize and analyze the relationship between two variables and describe their individual distributions on the same plot.  
- Compares two distributions and plots a scatter plot by default.  
- By default, the graphs on the either axes are histograms of each variables (columns)

```python
sns.jointplot(data=crash, x='speeding', y='alcohol', kind='hex') 
plt.show()

# kind='reg' for regression plot
# kind='kde' for kernel density estimation plot : estimates the density of the data
# kind='hex' for joint plot with hexagonal bins
```

#### Kernel Density Estimation Plot

- KDE plot depicts the probability density function of the continuous or non-parametric data variables  
    i.e. we can plot for the univariate or multiple variables altogether.

```python
sns.kdeplot(crash['alcohol'])
plt.show()
```

#### Pair Plot

- Pair plots plot the relationships across the entire dataframe's numerical values.  
- Basically a scatter plot matrix.

```python
sns.pairplot(data=crash, kind='reg')
plt.show()

# kind='reg' : regression plot, the default value is 'scatter'
```

```python
sns.pairplot(data=tips, hue='sex', palette='bright')
plt.show()
```

#### Rug Plot

- A rug plot is a plot of data for a single quantitative variable, displayed as marks along an axis.  
- It is used to visualise the distribution of the data.   
- As such it is analogous to a histogram with zero-width bins, or a one-dimensional scatter plot.  
- Displays the individual density distributions of two variables with ticks on either axes.  
- Used to complement other plots.

```python
sns.rugplot(data=tips, x='total_bill', y='tip', hue='sex')
plt.show()
```

## Categorical Plots

- Categorical plots are basically used for visualizing the relationship between variables.  - 
- Those variables can be either be completely numerical or a category like a group, class or division.  
- These kinds of plots allow us to choose a numerical variable, like age and plot the distribution of age for each category in a selected categorical variable.

#### Bar Plot

- A bar plot or bar chart is a graph that represents the category of data with rectangular bars with lengths and heights that is proportional to the values which they represent.  
- The bar plots can be plotted horizontally (column graph) or vertically (bar graph).  
- A bar chart describes the comparisons between the discrete categories.  
- One of the axis of the plot represents the specific categories being compared, while the other axis represents the measured values corresponding to those categories.

```python
sns.barplot(data=tips, x='sex', y='total_bill', orientation='vertical')
# estimator=np.mean, np.median, np.std 
# estimator is used to assign the type of data to for the bars
plt.show()
```

#### Count Plot

- A count plot can be thought of as a histogram across a categorical, instead of quantitative variable.
- Shows the counts of observations in each categorical bin using bars.

```python
sns.countplot(data=tips, x='sex', hue='sex')
# no need to assign hue here, I specified hue just to display the legend in my plot
plt.show()
```

#### Box Plot

- Box Plot is a method for graphically demonstrating the locality, spread and skewness groups of numerical data through their quartiles.
- A box and whisker plot displays the five-number summary of a set of data and outliers. 
- The five-number summary is the minimum, first quartile, median, third quartile, and maximum.
- Draws a box from the first quartile to the third quartile. A vertical line goes through the box at the median.


```python
tips['day'].unique()

plt.figure(figsize=(10, 5))
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex', palette='rocket')
plt.legend(loc=2)
plt.show()
```

#### Violin Plot

- A violin plot is a hybrid of a box plot and a kernel density plot, which shows peaks in the data.
- It is used to visualize the distribution of numerical data.
- Unlike a box plot that can only show summary statistics, violin plots depict summary statistics and the density of each variable.

```python
plt.figure(figsize=(10, 5))
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex', split=True, palette='dark')
# split=True combines two categories of hues and compares the density between them
plt.show()
```

#### Strip Plot

- A strip plot is a single-axis scatter plot that is used to visualise the distribution of many individual one-dimensional categorical values.
- The values are plotted as dots along one unique axis, and the dots with the same value can overlap showing the density.

```python
plt.figure(figsize=(10, 5))

sns.stripplot(data=tips, x='day', y='total_bill', palette='dark', jitter=True, hue='sex', dodge=True)
# jitter:True -> to seperate the points even more
# dodge:True -> completely  seperate the points for each categorical value in hue
plt.show()
```

#### Swarm Plot

- A swarm plot is a type of scatter plot that is used for representing categorical values.
- It is very similar to the strip plot, but it avoids the overlapping of points.
- Usually used with other plots.
- In our example, we used swarm plot to show the density distribution for the data in a violin plot.

```python
sns.violinplot(data=tips, x='day', y='total_bill')
sns.swarmplot(data=tips, x='day', y='total_bill', color='white')
plt.show()
```

## Matrix Plots

- A matrix diagram compares groups of data within a larger category of information. 
- This type of data visualization shows how these groups interact with and influence each other.
- Helps to simultaneously explore the associations between thousands of subjects, variables, and their interactions, without needing to first reduce the dimensions of the data.

#### Heatmaps

- A heat map is a two-dimensional representation of data in which variables are represented by colors.

```python
plt.figure(figsize=(10, 8))
sns.set_context('paper', font_scale=1.2)

sns.heatmap(crash.corr(), annot=True, cmap='Blues')

plt.show()
```

```python
# flights = flights.pivot_table(index='month', columns='year', values='passengers')
flights

sns.heatmap(data=flights, cmap='Purples', linecolor='white', linewidth=0.5)
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.show()
```

#### Cluster Map

- Clustered column charts are a type of graph used to compare data sets. 
- They are perfect for visualizing data that is grouped into categories. 
- Clustered column charts are often used to compare sales figures or other data that change over time.
- It is basically a hierachically clustered heatmap.
- A heatmap doesnot look for cluster pattern in the dataset and simply works with what it has.
- But a clustermap looks for cluster patterns and groups them according, thus the x-axis order might be altered.

```python
iris.pop('species') # remove 'species' column to develop a cluster map (as it only processes numerical values).

sns.clustermap(data=iris)
plt.show()
```

```python
sns.clustermap(flights, cmap='Blues', standard_scale=1)
plt.show()
```

#### Pair Grid

- Pair Grid allows us to draw a grid of subplots using the same plot type to visualize data.
- Unlike FacetGrid, it uses different pair of variable for each subplot.
- It forms a matrix of sub-plots.
- It is also sometimes called as scatterplot matrix.

```python
iris = sns.load_dataset('iris') # we popped the species column so reinitialing the dataframe

iris_grid = sns.PairGrid(data=iris, hue='species')

# iris_grid.map(plt.scatter)
iris_grid.map_diag(plt.hist)
# iris_grid.map_offdiag(plt.scatter)
iris_grid.map_upper(plt.scatter)
iris_grid.map_lower(sns.kdeplot)

plt.show()
```

```python
iris_grid_new = sns.PairGrid(data=iris, hue='species', 
                         y_vars=['sepal_length', 'sepal_width'], 
                        x_vars=['petal_length', 'petal_width'])

iris_grid_new.map(plt.scatter)
iris_grid_new.add_legend()

plt.show()
```

## Regression Plots

- Regression Plot creates a regression line between 2 parameters and helps to visualize their linear relationships.
- The regression line is an attempt to find the best fit through the points in the scatter plot.
- A positive coefficient indicates that as the value of the independent variable increases, the mean of the dependent variable also tends to increase.
- A negative coefficient suggests that as the independent variable increases, the dependent variable tends to decrease.

```python
plt.figure(figsize=(10, 8))
sns.set_context('paper', font_scale=1.2)

sns.lmplot(data=tips, x='total_bill', y='tip', hue='sex', 
           markers=['o', '*'], scatter_kws={'s':100, 'linewidth':0.5, 'edgecolor':'black'})

plt.show()
```

```python
sns.lmplot(data=tips, x='total_bill', y='tip', col='sex', row='time', height=5, aspect=1)

plt.show()
```
